# KMeans Clustering

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/jacobjoe/data_for_jupyter_notebook/master/iris_data_csv.csv')
print(df.head())
print(df.shape)
```

```python
plt.rcParams['figure.figsize'] = (20, 10)
plt.scatter(df['PetalLengthCm'], df['PetalWidthCm'], c = 'red')
plt.xlabel('Length')
plt.ylabel('Width')
```

```python
km = KMeans(n_clusters=3)
y_predict = km.fit_predict(df[['PetalLengthCm', 'PetalWidthCm']])
y_predict
```

```python
df['cluster'] = y_predict
df.head()
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]

plt.scatter(df1['PetalLengthCm'], df1['PetalWidthCm'], c = 'red', label = 'Petal')
plt.scatter(df2['PetalLengthCm'], df2['PetalWidthCm'], c = 'blue', label = 'Petal')
plt.scatter(df3['PetalLengthCm'], df3['PetalWidthCm'], c = 'green', label = 'Petal')
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], c = 'black', marker = '*', label = 'Centroid')

plt.xlabel('Length in cm')
plt.ylabel('Width in cm')
plt.legend()
plt.show()
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['PetalLengthCm']])
df['PetalLengthCm'] = scaler.transform(df[['PetalLengthCm']])

scaler.fit(df[['PetalWidthCm']])
df['PetalWidthCm'] = scaler.transform(df[['PetalWidthCm']])
```

```python
df.head()
```

```python
km = KMeans(n_clusters=3)
y_predict = km.fit_predict(df[['PetalLengthCm', 'PetalWidthCm']])
y_predict
```

```python
df['cluster'] = y_predict
df.head()
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]

plt.rcParams['figure.figsize'] = (20, 10)
plt.scatter(df1['PetalLengthCm'], df1['PetalWidthCm'], c = 'red', label = 'Petal')
plt.scatter(df2['PetalLengthCm'], df2['PetalWidthCm'], c = 'blue', label = 'Petal')
plt.scatter(df3['PetalLengthCm'], df3['PetalWidthCm'], c = 'green', label = 'Petal')
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], c = 'black', marker = '*', label = 'Centroid')

plt.xlabel('Length in cm')
plt.ylabel('Width in cm')
plt.legend()
plt.show()
```

# Elbow Plot

```python
sse = []
k_rng = range(1, 10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['PetalLengthCm', 'PetalWidthCm']])
    sse.append(km.inertia_)
```

```python
plt.rcParams['figure.figsize'] = (20, 10)
plt.xlabel('Length in cm')
plt.ylabel('Width in cm')
plt.plot(k_rng, sse, c = 'red')
```
